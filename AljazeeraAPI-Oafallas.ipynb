## Data Collection: News Site
This notebook collects data on news articles from Aljazeera from March 11, 2021 to March 12, 2021 through API access via NewsAPI

### Import Package

```python
import requests
```

### Make API request with appropriate parameters
NewsAPI limits free accounts to only first 100 results in a query, so the query will be divided into the 2 dates which have less than 100 results each to get all articles

```python
url = 'https://newsapi.org/v2/everything?'
api_key = '620cb58c82704c31b42ef4a105ddbe45'
parameters_1 = {
    'sources': 'al-jazeera-english',
    'language': 'en',
    'from': '2021-3-11',
    'to': '2021-3-11',
    'sortBy': 'publishedAt',
    'pageSize': '100',
    'page': '1',
    'apiKey': api_key
}

response_1 = requests.get(url, params = parameters_1)

parameters_2 = {
    'sources': 'al-jazeera-english',
    'language': 'en',
    'from': '2021-3-12',
    'to': '2021-3-12',
    'sortBy': 'publishedAt',
    'pageSize': '100',
    'page': '1',
    'apiKey': api_key
}

response_2 = requests.get(url, params = parameters_2)
```

### Display result

```python
response_1_json = response_1.json()
response_1_json
```

```python
response_2_json = response_2.json()
response_2_json
```

### Isolate articles and Combine Articles from March 11 and March 12

```python
response_1_json.keys()
```

```python
articles_1_json = response_1_json['articles']
articles_2_json = response_2_json['articles']
articles_json = articles_1_json + articles_2_json
len(articles_json)
```

```python
def format_output(articles):
    formatted_articles = []
    
    for i in range(len(articles)):
        article_dict = {
            "date": articles[i]['publishedAt'],
            "title": articles[i]['title'],
            "full_article": articles[i]['content'],
            "author": articles[i]['author']
        }
        formatted_articles.append(article_dict)
        
    return formatted_articles
```

```python
articles_formatted_json = format_output(articles_json)
articles_formatted_json
```

### Export to JSON file

```python
import json
with open('aljazeera-articles.json', 'w') as outfile:
        json.dump(articles_formatted_json, outfile)
```

### Verify JSON file

```python
with open('aljazeera-articles.json') as json_file:
        data = json.load(json_file)
len(data)
```
